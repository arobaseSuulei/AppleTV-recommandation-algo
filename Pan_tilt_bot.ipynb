```python
import gymnasium as gym
from gymnasium import spaces
import numpy as np

class PanTiltEnv(gym.Env):
    def __init__(self):
        super().__init__()

        self.observation_space = spaces.Box(
            low=np.array([-1, -1, -90, -45]),
            high=np.array([1, 1, 90, 45]),
            dtype=np.float32
        )

        self.action_space = spaces.Box(
            low=np.array([-1, -1]),
            high=np.array([1, 1]),
            dtype=np.float32
        )

        self.pan_angle = 0.0
        self.tilt_angle = 0.0
        self.ball_x = 0.0
        self.ball_y = 0.0
```
